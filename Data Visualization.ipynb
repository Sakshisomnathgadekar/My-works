```python
## Box Plot       inner part is called IQR inter Quator Range
```

```python
marks = [67,63,57,59,70,55,68,53,95,23]
```

```python
plt.boxplot(marks);
```

```python
q3 = np.percentile(marks,75)
q1 = np.percentile(marks,25)
q3,q1
```

```python
iqr = q3 - q1
iqr
```

```python
ub = q3+(1.5*iqr)
lb = q1-(1.5*iqr)
```

```python
ub,lb
```
